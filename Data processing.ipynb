# Skewness checking

```python
from scipy.stats import skew
def skewedness(data):
    skewed=[]
    for col in data:
        sk=skew(data[col])
        if(sk>=1.96 or sk<=-1.96):
            skewed.append(col)
    return skewed

print(skewedness(data_num))
```

```python
data['YearsSinceLastPromotion'].skew()
```

###### Skewness tells us about wheteher the perticular feature is distributed to the left or right or its distributed normally. The skewness value should be ranges between -1.96 to 1.96, oherwise the feature is skewed.
###### From the above code we can see that 'YearsSinceLastPromotion' column is skewed

## Handling skewness

```python
data['YearsSinceLastPromotion']=np.sqrt(data['YearsSinceLastPromotion'])
data['YearsSinceLastPromotion'].skew()
```

###### Here we are converting the skewed data by taking square root transformation

# Feature selection

###### Feature selection is the method where we select  those features which contribute most to the our predict(dependent) variable. Having irrelavant features in our data can decrease the accuracy of the models and make our model learn based on irrelavant features.

## Feature selection by KBest subset method

```python
# Feature extraction
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
  
test = SelectKBest(score_func=chi2, k=15)  
fit = test.fit(X, Y)

dfscores=pd.DataFrame(fit.scores_,columns=["Score"]) #Getting the scores
dfcolumns=pd.DataFrame(X.columns,columns=["features"]) #Getting the columns names
feature_rank=pd.concat([dfcolumns,dfscores],axis=1) #getting both column name and scores together
print(feature_rank)
```

###### Using Kbest subset method we are going tom select the kbest features by considring the scores of the features.
###### Above we've displayed the the each feature with its score

```python
feature_rank.columns = ['feature_names','Scores']
print(feature_rank.nlargest(20,'Scores'))
```

# Handling imbalance

```python
y.value_counts()
```

###### from the value count of target variable we can see that the dataset is imablance. If the dataset is imbalance our output will going to be bias towards the majority class, therefor we need to handle the imbalance.

### Handling imbalance using SMOTE

```python
from imblearn.over_sampling import SMOTE
oversampling=SMOTE()
X_smote,Y_smote=oversampling.fit_sample(x,y)
```

###### Here we've applied smote method

###### SMOTE first selects a  minority class instance at  random and find its k nearest minority class neighbours. Then it will synthetically create the observations by finding the mean, untill the minority class observation reaches to majoruty class.

```python
print(X_smote.shape)
print(Y_smote.shape)
```
